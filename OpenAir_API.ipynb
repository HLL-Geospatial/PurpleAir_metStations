```python
import openaq
import pandas as pd
```

```python
api = openaq.OpenAQ()
```

```python
status, resp = api.measurements(country = 'US', parameter = 'pm25', include_fields = ['attribution', 'sourceName'], limit = 10000)
```

```python
resp['results']
```

```python
data = resp['results']
```

```python
data
```

```python
coordinates = [(entry['coordinates']['latitude'], entry['coordinates']['longitude']) for entry in data]
```

```python
df = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
```
